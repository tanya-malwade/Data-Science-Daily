***Today’s Goal:*** Understand:
                    - How orders connect to customers
                    - What is a key
                    - Why we don’t merge blindly

#### Import packages

```python
import pandas as pd
```

#### Load TWO datasets (Only two)

```python
orders_df = pd.read_csv("DATA/olist_orders_dataset.csv")
customers_df = pd.read_csv("DATA/olist_customers_dataset.csv")
```

```python
orders_df.head()
```

```python
customers_df.head()
```

Which column appears in both?

#### Find the connecting column
- In orders_df → customer_id
- In customers_df → customer_id

#### Check if the key is unique

```python
customers_df['customer_id'].nunique()
```

```python
len(customers_df)
```

##### unique customer_id's and len of customer_id's both are same means all customer_id's are unique 
- If not equal → customer can place multiple orders (normal).

##### Relationship Understanding
- One customer can have many orders
- orders_df.customer_id → foreign key
- customers_df.customer_id → primary key

#### Check missing customer_id 

```python
orders_df['customer_id'].isna().sum()
```

- Missing key = broken relationship.

#### Conclusion 
- orders and customers are connected using customer_id
- One customer can place multiple orders
- customer_id is unique in customers dataset
- No missing customer_id found in orders
- Merging requires understanding relationship first

#### Learning 
- We check the relationship between two tables to correctly connect related data, avoid mismatches, and get accurate results when combining information.
- Checking relationships between tables ensures data integrity and helps choose the correct join method for accurate analysis and reporting.

***Why we join the tables?***

We join tables to combine related information from different places so we can understand the full story and answer real-world questions.

### DAY 6 — UNDERSTANDING & FIXING DATA TYPES
***Today’s Goal:***
- What data types are 
- Why wrong data types are dangerous
- How to convert data types safely (only dates today)

### Load merged dataset

```python
import pandas as pd

orders_df = pd.read_csv("DATA/olist_orders_dataset.csv")
customers_df = pd.read_csv("DATA/olist_customers_dataset.csv")

merged_df = pd.merge(
    orders_df,
    customers_df,
    on='customer_id',
    how='left'
)
merged_df.head()
```

### Check data types

```python
merged_df.dtypes
```

### Identify date columns

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
```

```python
merged_df[date_cols].dtypes
```

- object → means text, not date

### Convert text into dates

```python
for col in date_cols:
    merged_df[col] = pd.to_datetime(merged_df[col],errors='coerce')
```

```python
merged_df[date_cols].dtypes
```

- errors='coerce' → if bad value, make it NaT (safe missing)

### Check if missing values changed

```python
merged_df[date_cols].isna().sum()
```

#### Conclusion

- Date columns were stored as text initially
- Converted date columns to datetime format safely
- Used errors='coerce' to avoid crashes
- Some invalid date values became missing after conversion


#### Learning Points 

- Data type tells the computer what kind of data it is
- Dates written as text cannot be used for calculations
- Real-world data often has wrong data types
- Fixing data types is important before any analysis

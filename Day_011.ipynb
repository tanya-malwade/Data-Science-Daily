***title*** (1 missing )
- one title missing here 
- we can remove these row 

```python
# 3. Drop row where title is missing
df = df.dropna(subset=['title'])
```

***description*** (18 missing)
- 18 rows missing the description 
- not imp in EDA but 
- u can fill it with 'No Description'

```python
# 1. Fill age_certification
df['description'] = df['description'].fillna('No Description')
```

***seasons*** (3759 missing)
- movies not have seasons 
- movies = missing sessions
- we can fill it using 0

```python
# 2. Fill seasons (movies = 0)
df['seasons'] = df['seasons'].fillna(0)
```

***age_certification*** (2610 missing)
- these are the missing age groups 
- if movies or serirs are so old or international containt 
- we can fill it using 'Not Rated' using fill na

```python
# 1. Fill age_certification
df['age_certification'] = df['age_certification'].fillna('Not Rated')
```

***imdb_id*** (444 missing) but 
- we don't use id column in analysis 
- it is only use for identity 
- so we can drop this column 

```python
df.drop(columns=['imdb_id'], inplace=True)
```

***imdb_score*** (522 Missing)
- means IMDb rating (0â€“10)
- not used in EDA but can be imp for ML
- we can fill it using median of imdb_score

```python
df['imdb_score'].fillna(df['imdb_score'].median(), inplace=True)
```

***imdb_votes*** (538 missing)
- use to understand the popularity 
- not used in EDA but important for ML
- fill with median 

```python
df['imdb_votes'].fillna(df['imdb_votes'].median(), inplace=True)
```

- both are numerical colums we can fill it with median

```python
df['tmdb_popularity'].fillna(df['tmdb_popularity'].median(), inplace=True)
df['tmdb_score'].fillna(df['tmdb_score'].median(), inplace=True)
```

#### Learnings
- Real-world datasets are often messy and require careful data cleaning.
- Missing values are not always errors; sometimes they provide useful information (e.g., missing seasons indicating a movie).
- Data cleaning involves logical decision-making, not just deleting rows or columns.
- Median is more reliable than mean for filling missing values in skewed numerical data.
- Understanding the meaning of each column is the foundation of effective exploratory data analysis.

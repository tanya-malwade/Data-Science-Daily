#### Import packages

#### Read the CSV files

##### Assumptions Before Merge

- orders.customer_id connects to customers.customer_id
- One customer can have multiple orders
- customers table has one row per customer

```python
customers_df.shape
```

#### Perform SAFE merge

```python
merged_df = pd.merge(
    orders_df,
    customers_df,
    on='customer_id',
    how='left'
)
```

LEFT JOIN:
- Take everything from the left table
- Add matching data from the right table
- If no match → keep left data, fill right with empty (NaN)

#### Check rows AFTER merge

#### Conclusion

- Merge completed without increasing row count
- Left join preserved all orders
- Customer details successfully attached
- No unexpected row duplication noticed

#### Learnings:
- We merge columns to bring related information together in one table so analysis, reporting, and decision-making become easy and accurate.
- customer_id is used as the join key because it uniquely identifies customers and exists in both tables, ensuring accurate and reliable data merging.
- A left join is used to preserve all rows from the primary table while adding matching data from the secondary table, ensuring no critical records are lost.
